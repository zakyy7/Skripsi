```python
# --- KONFIGURASI UTAMA ---
project_path = '/content/drive/MyDrive/pothole_project'  # Folder penyimpanan drive
run_name = 'train_final_pothole'                            # Nama folder eksperimen
data_yaml = '/content/Road-Damage-Detection-4/data.yaml'

# Cek apakah ada checkpoint terakhir (last.pt)
last_ckpt = os.path.join(project_path, run_name, 'weights', 'last.pt')
resume_training = False

if os.path.exists(last_ckpt):
    print(f"⚠️ Ditemukan checkpoint training sebelumnya di: {last_ckpt}")
    print("🔄 Melanjutkan training (Resume)...")
    model = YOLO(last_ckpt)
    resume_training = True
else:
    print("🆕 Memulai training baru dari awal dengan YOLO11s...")
    model = YOLO('yolo11s.pt') # Menggunakan versi Small sesuai request

# --- MULAI TRAINING ---
# Jika resume=True, parameter lain akan otomatis diambil dari save-an sebelumnya
if resume_training:
    model.train(resume=True)
else:
    model.train(
        data=data_yaml,
        epochs=100,
        patience=30,

        # Optimasi Hardware (Batch kecil karena img 832)
        batch=8,        # Turunkan ke 8 atau 4 jika imgsz=832 agar tidak error memori
        imgsz=832,      # Resolusi tinggi untuk deteksi lubang kecil/jauh

        # Strategi Anti False Positive (Batu/Bayangan)
        optimizer='AdamW',
        lr0=0.001,
        hsv_h=0.015,
        hsv_s=0.7,      # Saturation tinggi (bedakan aspal basah/kering)
        hsv_v=0.4,      # Value variatif (tahan bayangan pohon)

        # Augmentasi
        degrees=10.0,
        translate=0.1,
        scale=0.5,
        mosaic=1.0,
        mixup=0.1,

        # Lokasi Simpan
        project=project_path,
        name=run_name,
        exist_ok=True   # Izinkan menimpa folder jika resume gagal manual
    )

print("✅ Proses selesai/berhenti. Cek folder Drive Anda.")
```
